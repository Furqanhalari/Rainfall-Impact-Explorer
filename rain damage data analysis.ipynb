```python
flood_df = df[
    df["Disaster Type"].str.contains("Flood", case=False, na=False)
]
```

```python
flood_df
```

```python
flood_df = flood_df[[
    "Start Year",
    "Total Deaths",
    "No. Injured",
    "No. Affected",
    "No. Homeless",
    "Total Damage ('000 US$)"
]]
```

```python
flood_df
```

```python
flood_df = flood_df.rename(columns={"Start Year": "Year"})
```

```python
flood_df
```

```python
flood_df = flood_df.groupby("Year", as_index=False).sum()
```

```python
flood_df
```

```python
flood_df = flood_df.query("Year != 2025")
```

```python
flood_df
```

```python
flood_df
```

```python
flood_df
```

```python
flood_df = flood_df.drop(columns=["Total Damage ('000 US$)", "Total_Damage_USD"])
```

```python
flood_df
```

```python
flood_df["Total_Damage_Billion_PKR"] = (flood_df["Total_Damage_PKR"] / 1e9).round(2)
```

```python
flood_df
```

```python
flood_df.columns = flood_df.columns.str.strip()
```

```python
print(flood_df.columns)
```

```python
flood_df.head()
```

```python
flood_df = flood_df.drop(columns=["Total_Damage_Billion_PKR", "Total_Damage_Display_PKR"], errors="ignore")
```

```python
flood_df.head()
```

```python
flood_df["Total_Damage_Display"] = flood_df["Total_Damage_Display"].str.replace(" PKR", "", regex=False)
```

```python
flood_df.head()
```

```python
flood_df = flood_df.rename(columns={"Total_Damage_Display": "Total_Damage_Display_PKR"})
```
